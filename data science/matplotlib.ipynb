```python
x= np.arange(1,10)
y= x**2

plt.plot(x,y,'r') #o,.,^,*,--,-
plt.title('Squares')
plt.xlabel('x')
plt.ylabel('x^2')
plt.xlim(-5,15)
plt.ylim(-10,100)
plt.text(6,30,'Hello')
plt.grid()
plt.show()
# r for red color
#o is for cicle
```

```python
x= np.arange(-10,20)
y=x**2
plt.plot(x,y,'b--',label='quadratic')

x2= np.arange(-10,20)
y2=x2++ 90

plt.plot(x2,y2,'r',label='linear',linewidth=4)
plt.legend()

plt.show()
```

## Scatter 

```python
x= np.random.randint(1,100,10)
y=np.random.randint(1,100,10)

plt.scatter(x,y,color='y')
plt.scatter(y,x,color='g')
plt.show()
```

## Bar Chart

```python
x=np.array ([1,2,3,4])
plt.bar(x-0.1, [10,20,40,30], width=.2,label='BJP',color='r')
plt.bar(x+0.1, [20,30,10,40], width=.2,label='AAP',color='b')
plt.legend()
plt.show()
```

```python
#sigmoid(x) or logistic function = 1/1+e^-x
import math
sigmoid = lambda x: 1/(1+ math.exp(-x))
x=np.arange(-10,11)
y= [sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```
