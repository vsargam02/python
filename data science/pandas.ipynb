```python
weather = pd.read_csv('./data/weather.csv')
weather.head()
```

```python
weather_melt =pd.melt(weather,
               id_vars=['year','month','element'],
               var_name='day',
               value_name='temp')
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(index=['year','month','day'],
                        columns = 'element',
                        values='temp').reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('./output/weather_tidy.csv')
```

```python
def mean_func(arr):
    return sum(arr)/len(arr)
    
```

```python
weather_tidy[['month','year']].apply(mean_func,axis=1).head()
```

```python
def extractDay(row):
    val = row[0]
    return int(val[3:])
```

```python
weather_tidy[['day']]= weather_tidy[['day']].apply(extractDay,axis=1)
```
